```python
import numpy as np

def preprocess_input_model1(input_df):
    count = 0
    features = ['flow_duration', 'Header_Length', 'Protocol Type', 'Duration', 'Rate',
       'Srate', 'syn_count', 'urg_count', 'rst_count', 'HTTPS', 'Tot sum',
       'Min', 'Max', 'AVG', 'Std', 'Tot size', 'IAT', 'Number', 'Magnitue',
       'Radius', 'Covariance', 'Variance', 'Weight', 'LLC', 'SMTP', 'IRC',
       'Telnet', 'fin_count', 'rst_flag_number', 'syn_flag_number',
       'ece_flag_number', 'psh_flag_number']
    
    # Dictionary for mapping similar feature names
    similar_features_dict = {
        'flow_duration': ['Duration', 'Duration'],
        'Protocol Type': ['Protocol_encoded', 'Protocol_encoded'],
        'Duration': ['Duration', 'Duration'],
        'Rate': ['paket_rate'],
        'Srate': ['byte_rate'],
        'fin_count': ['FIN or RST']
        # Add more mappings as needed
    }
    
    # Initialize an array with zeros for all features
    input_array = np.zeros(len(features))
    
    # Fill in the values for the features present in the input_data
    for i, feature in enumerate(features):
        if feature in input_df.columns:
            input_array[i] = input_df[feature].values
            count += 1
        else:
            # Check if feature name has a mapping in the dictionary
            if feature in similar_features_dict:
                mapped_features = similar_features_dict[feature]
                for mapped_feature in mapped_features:
                    if mapped_feature in input_df.columns:
                        input_array[i] = input_df[mapped_feature].values
                        break  # Break out of loop once a matching feature is found
    
    return input_array





def preprocess_input_model2(input_df):
    count=0
    features = ['Timestamp', 'Scr_port', 'Des_port', 'Duration', 'Scr_bytes',
       'Des_bytes', 'Conn_state', 'is_syn_only', 'is_with_payload',
       'FIN or RST', 'Scr_pkts', 'Scr_ip_bytes', 'Des_pkts', 'Des_ip_bytes',
       'anomaly_alert', 'total_bytes', 'total_packet', 'paket_rate',
       'byte_rate', 'Scr_packts_ratio', 'Des_pkts_ratio', 'Scr_bytes_ratio',
       'Des_bytes_ratio', 'Avg_user_time', 'Avg_nice_time', 'Avg_system_time',
       'Avg_ideal_time', 'Avg_num_cswch/s', 'std_num_cswch/s',
       'OSSEC_alert_level', 'Login_attempt', 'Process_activity',
       'read_write_physical.process', 'Protocol_encoded', 'Service_coap',
       'Service_dns', 'Service_modbus', 'Service_mqtt', 'Service_other',
       'Service_echo', 'missed_bytes', 'File_activity', 'Avg_kbmemused',
       'is_privileged', 'Service_ssh', 'Service_imap', 'Std_ideal_time',
       'Service_mysql', 'Service_simple_service_discovery', 'Service_http',
       'Std_wtps', 'Service_websocket', 'is_SYN_with_RST', 'Avg_wtps',
       'Std_system_time', 'Succesful_login', 'Avg_rtps']
    # Initialize an array with zeros for all features
    input_array = np.zeros([len(features)])
    
    # Function to calculate number of mismatch characters
    def num_mismatch_chars(s1, s2):
        mismatch_count = 0
        min_len = min(len(s1), len(s2))
        for i in range(min_len):
            if s1[i] != s2[i]:
                mismatch_count += 1
        mismatch_count += abs(len(s1) - len(s2))  # Consider remaining characters as mismatches
        return mismatch_count
    
    # Fill in the values for the features present in the input_data
    for i, feature in enumerate(features):
        feature_lower = feature.lower().replace(" ", "").replace("-", "").replace("_", "")  # Convert to lowercase and remove underscores
        if feature in input_df.columns:
            count+=1
            input_array[i] = input_df[feature].values
        else:
            # Check if feature name matches with any regular expression mapping
            for input_key in input_df.columns:
                input_key_lower = input_key.lower().replace(" ", "").replace("-", "").replace("_", "")  # Convert to lowercase and remove underscores
                if num_mismatch_chars(feature_lower, input_key_lower) < 3:  # Threshold for mismatch characters
                    input_array[i] = input_df[input_key].values
                    count+=1
                    break
        
    if(count>=((1/3)*len(features))):
        return input_array
    else:
        return "None"
    

def preprocess_input_model3(input_df):
    count=0
    features = ['Src Port', 'Dst Port', 'Protocol', 'Flow Duration', 'Tot Bwd Pkts',
       'TotLen Fwd Pkts', 'TotLen Bwd Pkts', 'Fwd Pkt Len Max',
       'Fwd Pkt Len Mean', 'Bwd Pkt Len Max', 'Bwd Pkt Len Mean',
       'Bwd Pkt Len Std', 'Flow IAT Max', 'Fwd IAT Tot', 'Fwd IAT Mean',
       'Fwd IAT Max', 'Fwd IAT Min', 'Bwd IAT Tot', 'Bwd IAT Mean',
       'Bwd IAT Max', 'Bwd IAT Min', 'Fwd Header Len', 'Bwd Header Len',
       'Fwd Pkts/s', 'Bwd Pkts/s', 'Pkt Len Std', 'FIN Flag Cnt',
       'ACK Flag Cnt', 'Down/Up Ratio', 'Fwd Seg Size Avg', 'Subflow Fwd Byts',
       'Subflow Bwd Byts', 'Init Fwd Win Byts', 'Init Bwd Win Byts',
       'Fwd Act Data Pkts', 'Fwd Seg Size Min', 'Idle Mean', 'Idle Max',
       'Idle Min', 'Pkt Len Var', 'Flow IAT Min', 'Pkt Len Mean',
       'Flow IAT Std', 'Bwd Pkt Len Min', 'RST Flag Cnt', 'Bwd Byts/b Avg',
       'URG Flag Cnt', 'Flow Byts/s', 'Fwd Pkt Len Std', 'Active Max',
       'Bwd IAT Std', 'Bwd URG Flags', 'CWE Flag Count', 'Pkt Len Min',
       'ECE Flag Cnt', 'Bwd Pkts/b Avg', 'Pkt Size Avg', 'Flow Pkts/s',
       'Active Std', 'Bwd Blk Rate Avg', 'Fwd Pkt Len Min']
    # Initialize an array with zeros for all features
    input_array = np.zeros([len(features)])
    
    # Function to calculate number of mismatch characters
    def num_mismatch_chars(s1, s2):
        mismatch_count = 0
        min_len = min(len(s1), len(s2))
        for i in range(min_len):
            if s1[i] != s2[i]:
                mismatch_count += 1
        mismatch_count += abs(len(s1) - len(s2))  # Consider remaining characters as mismatches
        return mismatch_count
    
    # Fill in the values for the features present in the input_data
    for i, feature in enumerate(features):
        feature_lower = feature.lower().replace(" ", "").replace("-", "").replace("_", "")  # Convert to lowercase and remove underscores
        if feature in input_df.columns:
            count+=1
            input_array[i] = input_df[feature].values
            print(count)
        else:
            # Check if feature name matches with any regular expression mapping
            for input_key in input_df.columns:
                input_key_lower = input_key.lower().replace(" ", "").replace("-", "").replace("_", "")  # Convert to lowercase and remove underscores
                if num_mismatch_chars(feature_lower, input_key_lower) < 3:  # Threshold for mismatch characters
                    input_array[i] = input_df[input_key].values
                    count+=1
                    break
    
    if(count>=((1/3)*len(features))):
        return input_array
    else:
        return "None"
```
